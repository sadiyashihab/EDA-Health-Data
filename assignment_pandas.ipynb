Fill the null values by using interpolate

```python
df.fillna(df.interpolate(numeric_only=True).round(1))
print(df)
```

```python
# Correlation
df.corr()
```

Filtering using loc

```python
filtered_values = pn.where((df['Hours_per_day']>=5) & (df['Age']<= 40) & (df['Primary_streaming_service'].str.startswith('Y')))
print(filtered_values)
display(df.loc[filtered_values])
```

```python
filtere_values = pn.where((df['Hours_per_day']>=5) & (df['Age']<= 40) & (df['Anxiety']>=5)	&(df['Depression']>=5)&	(df['Insomnia']>=5)&	(df['OCD']>=5) & (df['Primary_streaming_service'].str.startswith('Y')))
print(filtere_values)
display(df.loc[filtere_values])
```

```python
df_merged = df.groupby('Primary_streaming_service')['Age'].apply(list).reset_index()
print(df_merged)
```
